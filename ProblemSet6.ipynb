```python
import pandas as pd
import altair as alt
pd.options.display.float_format = '{:.4f}'.format
from sklearn.neighbors import KNeighborsRegressor
```

```python
battingdf = pd.read_csv('Batting.csv')
salarydf = pd.read_csv('Salaries.csv')
```

My drafting bust is going to be Jeff Clement (clemeje01)

My drafting success is going to be Albert Pujols (pujolal01)

```python
#ID bust: clemeje01 Jeff clement
#ID success: pujolal01 Albert Pujols
```

```python
#Reduced data to after 1998, a few years before Pujols started playing
battingdf = battingdf.loc[battingdf['yearID'] >= 1998]
salarydf = salarydf.loc[salarydf['yearID'] >= 1998]
```

```python
battingdf.head()
```

```python
#Got the group averages for all columns
avgdf = battingdf.groupby(['yearID']).mean()
avgdf['playerID'] = 'Group Average'
avgdf=avgdf.reset_index()
```

```python
batclem = battingdf.loc[(battingdf['playerID'] == 'clemeje01')]
batbueh = battingdf.loc[(battingdf['playerID'] == 'pujolal01')]
combinedf = pd.concat([batclem, batbueh, avgdf], sort=True)
#Combined them all 
```

```python
avgsalarydf = salarydf.groupby(['yearID']).mean()
avgsalarydf['playerID'] = 'Group Average'
avgsalarydf=avgsalarydf.reset_index()
```

```python
salclem = salarydf.loc[(salarydf['playerID'] == 'clemeje01')]
salbueh = salarydf.loc[(salarydf['playerID'] == 'pujolal01')]
combinedf2 = pd.concat([salclem, salbueh, avgsalarydf], sort=True)
```

```python
combinedf2.head()
```

```python
alt.Chart(combinedf2).mark_point().encode(
    alt.X('yearID', scale=alt.Scale(zero=False)),
    y='salary',
    color='playerID',
).interactive()
```

The above chart is the player salaries. Pujols' salary starts well below average, but within a year is more than twice the average. Clement only has 1 entry for his salary, and it is quite a bit below average

```python
alt.Chart(combinedf).mark_point().encode(
    alt.X('yearID', scale=alt.Scale(zero=False)),
    y='SO',
    color='playerID',
).interactive()
```

Pujols strikes out a lot more than average, while Clement has a couple years above average and in a similar range as Pujols. 

```python
alt.Chart(combinedf).mark_point().encode(
    alt.X('yearID', scale=alt.Scale(zero=False)),
    y='RBI',
    color='playerID',
).interactive()
```

Despite striking out a similar amount as Pujols in 2008 and 2010, Clement's RBI is significantly lower than Pujols'. Clement's RBI hovered around average for a couple years, while Pujols has been doing very well.
